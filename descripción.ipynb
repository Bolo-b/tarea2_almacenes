# II. Dataset a limpiar: personas desaparecidas en México

**Dataset:** `RNPDNO-22-08-2023.csv`

Se trabajará con conjuntos de datos abiertos relacionados con personas desaparecidas en México, específicamente el Registro Nacional de Personas Desaparecidas y No Localizadas (RNPDNO).

```python
import pandas as pd

# Cargar el dataset con el encoding correcto para español
df = pd.read_csv('RNPDNO-22-08-2023.csv', low_memory=False, encoding='latin-1')

filas, columnas = df.shape

fechas = pd.to_datetime(df['Fecha de desaparición'], errors='coerce', dayfirst=True)
fecha_min = fechas.min()
fecha_max = fechas.max()

print(f"***Descripción del Dataset***")
print(f"Número de filas: {filas:,}")
print(f"Número de columnas: {columnas}")
print(f"Periodo cubierto: del {fecha_min.strftime('%d/%m/%Y')} al {fecha_max.strftime('%d/%m/%Y')}")
```

```python
Dada la ejecución del código anterior, tenemos la siguiente descripción del dataset:

* Número de filas: 116,945 registros.
* Número de columnas: 11 variables descriptivas.
* Periodo de fechas cubierto: Desde el 01/05/1968 hasta el 20/08/2023.
* Unidad de análisis: Cada fila representa un reporte individual de una persona desaparecida o no localizada.
                          Se identifica por un número de registro y se encuentra asociado a una autoridad.
```
